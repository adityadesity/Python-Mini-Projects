```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/14_naive_bayes/spam.csv')
df.head()
```

```python
df.groupby('Category').describe()
```

```python
c_dummies = pd.get_dummies(df['Category'])
```

```python
c_dummies
```

```python
df = pd.concat([df,c_dummies], axis = 'columns')
```

```python
df.head()
```

```python
df = df.drop(['Category'], axis = 'columns')
```

```python
df.head()#Errorenous
```

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/14_naive_bayes/spam.csv')
df1.head()
```

```python
df1['spam'] = df1['Category'].apply(lambda x: 1 if x=='spam' else 0)
df1.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df['Message'],df.spam , test_size=0.25, random_state=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()[:3]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
emails = {
    "Hey, Adi, can we meet together for a cricket match tomorrow?",
    "upto 20% discount on parking,exclusive offer just for you. Hurry up!"
}
emails_cnt = v.transform(emails)
model.predict(emails_cnt)
```

```python
eg = {
"Welcome to the August 5, 2022, edition of ACM TechNews, providing timely information for IT professionals three times a week",
"Hurry up! Excluxive offer tonight. upto 40% off!"
}
eg_cnt = v.transform(eg)
model.predict(eg_cnt)
```

```python
X_test_count = v.transform(X_test)
model.score(X_test_count,y_test)
```

```python
import pickle
with open('spamdetector.pkl','wb') as f:
    pickle.dump(model,f)
```
